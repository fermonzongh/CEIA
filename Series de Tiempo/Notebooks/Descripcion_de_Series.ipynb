````python
import pandas as pd
import matplotlib.pyplot as plt
````

# Serie 1

````python
url = "https://raw.githubusercontent.com/charlieromano/TimeSeries/main/Datasets/S1MME_week45.csv"
df_1 = pd.read_csv(url)
````

````python
df_1.head()
````

````python
df_1 = df_1.drop("NENAME", axis = 1)
````

````python
df_1.plot(x="MyDay",y="S1_mode_combined_attach_request_times_SEQ")
plt.show()
````

````python
df_1.describe()
````

````python
_,ax1=plt.subplots()
ax1.hist(df_1["S1_mode_combined_attach_success_times_SEQ"],bins=30)
````

````python
df_1.plot(x="MyDay",y="S1_mode_combined_attach_success_times_SEQ")
plt.show()
````

En la serie 1, columna "S1_mode_combined_attach_success_times_SEQ", tenemos 3 periodos a simple vista, completando para el de menor frecuencia 6 ciclos en el tiempo analizado, para la segunda frecuencia, completa 18 ciclos, esto lo vemos en los picos y valles de la serie, y para la mayor de las frecuencias, tenemos 36 ciclos aproximadamente. La tendencia no es creciente ni decreciente, tenemos una media aproximada de 231 401, y varianza 67 367, obtenidas mediante el metodo ```describe()```. El menor valor es 31 966 y el mayor valor es 469 172.
